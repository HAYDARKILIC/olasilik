## 1. Temel Sayma Kuralı

**Kural:** Bir deney $r$ farklı aşamadan oluşuyorsa ve $i$-inci aşamada $n_i$ olasılık varsa, toplam sonuç sayısı:

$$n_1 \times n_2 \times \cdots \times n_r = \prod_{i=1}^{r} n_i$$

### Örnek 1.1 — Takım Oluşturma
5 kız ve 2 erkekten oluşan bir gruptan 1 kız + 1 erkekten oluşan kaç farklı takım kurulabilir?

```python
kizlar  = ['G1', 'G2', 'G3', 'G4', 'G5']
erkekler = ['B1', 'B2']

# Temel sayma kuralı
toplam = len(kizlar) * len(erkekler)
print(f"Kız sayısı  : {len(kizlar)}")
print(f"Erkek sayısı: {len(erkekler)}")
print(f"Toplam takım: {len(kizlar)} × {len(erkekler)} = {toplam}")

print("\nTüm takımlar:")
takim_no = 1
takimlar = []
for k in kizlar:
    for e in erkekler:
        print(f"  {takim_no:2d}. ({k}, {e})")
        takimlar.append((k, e))
        takim_no += 1
```

```python
# Görselleştirme: Çarpım tablosu
fig, ax = plt.subplots(figsize=(9, 4))
ax.set_xlim(0, len(erkekler) + 0.5)
ax.set_ylim(0, len(kizlar) + 0.5)
ax.set_xticks(range(1, len(erkekler) + 1))
ax.set_yticks(range(1, len(kizlar) + 1))
ax.set_xticklabels(erkekler, fontsize=12)
ax.set_yticklabels(kizlar,   fontsize=12)
ax.set_xlabel("Erkekler", fontsize=12)
ax.set_ylabel("Kızlar",   fontsize=12)
ax.set_title(f"Temel Sayma Kuralı: {len(kizlar)} × {len(erkekler)} = {toplam} farklı takım",
             fontsize=13, fontweight='bold')
ax.grid(True, linestyle='--', alpha=0.4)

renkler = plt.cm.Set2(np.linspace(0, 1, len(takimlar)))
for idx, (k, e) in enumerate(takimlar):
    xi = erkekler.index(e) + 1
    yi = kizlar.index(k)   + 1
    ax.scatter(xi, yi, s=400, color=renkler[idx], zorder=3, edgecolors='black', linewidths=0.8)
    ax.text(xi, yi, f"{k}\n{e}", ha='center', va='center', fontsize=7, fontweight='bold')

plt.tight_layout()
plt.show()
```

### Örnek 1.2 — Plaka Sayısı (Genelleştirilmiş Kural)

```python
harf_sayisi  = 26   # A-Z
rakam_sayisi = 10   # 0-9

# Genelleştirilmiş çarpım kuralı: 26^3 × 10^3
asamalar = [harf_sayisi, harf_sayisi, harf_sayisi,
            rakam_sayisi, rakam_sayisi, rakam_sayisi]

toplam_plaka = math.prod(asamalar)

print("Aşama  | Olasılık Sayısı")
print("-" * 28)
isimler = ["1. Harf", "2. Harf", "3. Harf", "1. Rakam", "2. Rakam", "3. Rakam"]
for isim, n in zip(isimler, asamalar):
    print(f"{isim:<10} | {n}")
print("-" * 28)
print(f"\nToplam = 26 × 26 × 26 × 10 × 10 × 10")
print(f"       = 26³ × 10³")
print(f"       = {26**3:,} × {10**3:,}")
print(f"       = {toplam_plaka:,} farklı plaka")
```

## 2. Permütasyonlar

### 2.1 — Temel Permütasyon
$n$ farklı nesnenin **tüm** sıralamalarının sayısı:
$$P(n) = n!$$

### 2.2 — Tekrarlı Nesnelerle Permütasyon
$n$ nesne içinde $n_1$ benzer, $n_2$ benzer, …, $n_r$ benzer varsa:
$$P = \frac{n!}{n_1!\, n_2!\, \cdots\, n_r!}$$

### 2.3 — $n$ nesneden $r$ seçmenin permütasyonu
$$P(n, r) = \frac{n!}{(n-r)!}$$

```python
# Yardımcı fonksiyonlar
def faktoriyel(n):
    """n! hesaplar"""
    return math.factorial(n)

def perm(n, r=None):
    """n'den r seçmenin permütasyonu: P(n,r) = n! / (n-r)!"""
    if r is None:
        r = n
    if r > n:
        return 0
    return faktoriyel(n) // faktoriyel(n - r)

def tekrarli_perm(nesne_listesi):
    """Tekrarlı nesnelerle permütasyon: n! / (n1! * n2! * ...)"""
    n = len(nesne_listesi)
    sayac = Counter(nesne_listesi)
    payda = math.prod(faktoriyel(v) for v in sayac.values())
    return faktoriyel(n) // payda

print("Permütasyon fonksiyonları tanımlandı ✓")
```

### Örnek 2.1 — a, b, c Harflerinin Permütasyonları

```python
harfler = ['a', 'b', 'c']
n = len(harfler)

# Tüm permütasyonları listele
tum_permler = list(itertools.permutations(harfler))

print(f"n = {n} farklı harf → {n}! = {faktoriyel(n)} permütasyon")
print("\nTüm permütasyonlar:")
for i, p in enumerate(tum_permler, 1):
    print(f"  {i}. {''.join(p)}")

print(f"\nDoğrulama: itertools ile bulunan permütasyon sayısı = {len(tum_permler)}")
if len(tum_permler) == faktoriyel(3):
    print(f"Formül ile: 3! = {faktoriyel(3)}  ✓")
else:
    print(f"Formül ile: beklenen={faktoriyel(3)}, bulunan={len(tum_permler)}  ✗")
```

### Örnek 2.2 — "FACEBOOK" Kelimesinin Permütasyonları (Tekrarlı)

```python
kelime = list("FACEBOOK")
n = len(kelime)

print(f"Kelime: {''.join(kelime)}")
print(f"Harf sayısı (n): {n}")

# Harf frekansları
frekans = Counter(kelime)
print("\nHarf frekansları:")
for harf, adet in sorted(frekans.items()):
    print(f"  '{harf}': {adet} adet")

# Hesaplamalar
hepsi_farkli = faktoriyel(n)
tekrarli     = tekrarli_perm(kelime)

print(f"\n(a) Hepsi farklıymış gibi: 8! = {hepsi_farkli:,}")
print(f"(b) 'o' tekrarını düzeltince: 8! / 2! = {hepsi_farkli:,} / {faktoriyel(2)} = {tekrarli:,}")

# itertools ile doğrulama
gercek = len(set(itertools.permutations(kelime)))
print(f"\nDoğrulama (itertools unique perm sayısı): {gercek:,}")
print("Formül doğru ✓" if tekrarli == gercek else "Formül yanlış ✗")
```

## 3. Kombinasyonlar

**Sıralama önemli değil!** $n$ nesneden $r$ seçme yolu:

$$\binom{n}{r} = C(n,r) = \frac{n!}{(n-r)!\, r!}$$

**Permütasyon vs Kombinasyon:**
$$\binom{n}{r} = \frac{P(n,r)}{r!}$$

Çünkü seçilen $r$ nesnenin $r!$ farklı sıralaması birbirinin aynısı sayılır.

```python
def kombinasyon(n, r):
    """C(n,r) = n! / ( (n-r)! * r! )"""
    if r < 0 or r > n:
        return 0
    return math.comb(n, r)   # Python 3.8+ built-in

print("Kombinasyon örnekleri:")
for (n, r) in [(4,2), (6,2), (52,5), (10,3)]:
    c = kombinasyon(n, r)
    p = perm(n, r)
    print(f"  C({n},{r}) = {n}!/({r}!·{n-r}!) = {c:>10,}   [P({n},{r})={p:,} / {r}!={faktoriyel(r)}]")
```

### Örnek 3.1 — El Sıkışmaları: {1,2,3,4,5,6}'dan 2 Seçme

```python
kisiler = [1, 2, 3, 4, 5, 6]
n, r = 6, 2

kombinasyonlar = list(itertools.combinations(kisiler, r))

print(f"6 kişilik grupta el sıkışma sayısı = C(6,2) = {kombinasyon(n,r)}")
print(f"Formül: 6! / (4! · 2!) = {faktoriyel(6)} / ({faktoriyel(4)}·{faktoriyel(2)}) = {kombinasyon(n,r)}")
print(f"\nTüm el sıkışmaları:")
for i, (a, b) in enumerate(kombinasyonlar, 1):
    print(f"  {i:2d}. Kişi {a} ↔ Kişi {b}")

print(f"\nNot: Permütasyonda P(6,2)={perm(6,2)} seçenek vardı.")
print(f"El sıkışmada (1,2) = (2,1) olduğundan {perm(6,2)} / 2! = {kombinasyon(n,r)}")
```

```python
# El sıkışma ağ grafiği
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Sol: el sıkışma ağı
ax = axes[0]
n_kisiler = 6
acilar = np.linspace(0, 2 * np.pi, n_kisiler, endpoint=False)
x_pos = np.cos(acilar)
y_pos = np.sin(acilar)

for (i, j) in itertools.combinations(range(n_kisiler), 2):
    ax.plot([x_pos[i], x_pos[j]], [y_pos[i], y_pos[j]],
            'b-', alpha=0.3, linewidth=1.2)

renkler_node = plt.cm.Set1(np.linspace(0, 1, n_kisiler))
for i in range(n_kisiler):
    ax.scatter(x_pos[i], y_pos[i], s=500, color=renkler_node[i],
               zorder=5, edgecolors='black', linewidths=1.5)
    ax.text(x_pos[i] * 1.18, y_pos[i] * 1.18, f"Kişi {i+1}",
            ha='center', va='center', fontsize=10, fontweight='bold')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title(f"C(6,2) = {kombinasyon(6,2)} El Sıkışma",
             fontsize=12, fontweight='bold')

# Sağ: Permütasyon vs Kombinasyon karşılaştırması
ax2 = axes[1]
ns_range = range(2, 11)
r_sabit = 2
perm_vals = [perm(n, r_sabit) for n in ns_range]
comb_vals = [kombinasyon(n, r_sabit) for n in ns_range]

x = np.array(list(ns_range))
genislik = 0.35
ax2.bar(x - genislik/2, perm_vals, genislik, label=f'P(n,{r_sabit}) — Sıralama Önemli',
        color='#E07B54', alpha=0.85, edgecolor='black', linewidth=0.5)
ax2.bar(x + genislik/2, comb_vals, genislik, label=f'C(n,{r_sabit}) — Sıralama Önemsiz',
        color='#5B8DB8', alpha=0.85, edgecolor='black', linewidth=0.5)

ax2.set_xlabel('n (toplam eleman sayısı)', fontsize=11)
ax2.set_ylabel('Sayı', fontsize=11)
ax2.set_title(f'P(n,{r_sabit}) vs C(n,{r_sabit}) — r={r_sabit} için',
              fontsize=12, fontweight='bold')
ax2.legend(fontsize=9)
ax2.set_xticks(x)
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

### Örnek 3.2 — Poker Eli: 52 Karttan 5 Kart Seçme

```python
n_kart = 52
el_boyutu = 5

toplam_el = kombinasyon(n_kart, el_boyutu)
print(f"Toplam poker eli sayısı: C(52,5) = {toplam_el:,}")
print(f"Formül: 52! / (47! · 5!) = (52×51×50×49×48) / (5×4×3×2×1)")
print(f"      = {52*51*50*49*48:,} / {5*4*3*2*1} = {toplam_el:,}")

# Bazı özel el türleri
print("\n--- Özel Eller ---")

# Flush: Aynı renk 5 kart (4 renk × C(13,5))
flush = 4 * kombinasyon(13, 5)
print(f"Flush (aynı renk 5 kart)         : {flush:>8,}  → olasılık: {flush/toplam_el:.5f}")

# Full House: 3 aynı + 2 aynı  → C(13,1)×C(4,3) × C(12,1)×C(4,2)
full_house = kombinasyon(13,1)*kombinasyon(4,3) * kombinasyon(12,1)*kombinasyon(4,2)
print(f"Full House (3+2 aynı değer)      : {full_house:>8,}  → olasılık: {full_house/toplam_el:.5f}")

# Dörtlü: C(13,1)×C(4,4) × C(48,1)
dortlu = kombinasyon(13,1) * kombinasyon(4,4) * 48
print(f"Dörtlü (4 aynı değer)            : {dortlu:>8,}  → olasılık: {dortlu/toplam_el:.5f}")
```

### Kombinasyonların Özellikleri ve Binom Teoremi

```python
print("=" * 50)
print("Kombinasyon Özellikleri")
print("=" * 50)

# Özellik 1
for n in [5, 8, 10]:
    assert kombinasyon(n, 1) == n
    assert kombinasyon(n, n) == 1
print(f"✓ Özellik 1: C(n,1)=n, C(n,n)=1  →  Doğrulandı")

# Özellik 2: C(n,r) = C(n, n-r)
for n in [6, 9]:
    for r in range(n+1):
        assert kombinasyon(n, r) == kombinasyon(n, n-r)
print(f"✓ Özellik 2: C(n,r) = C(n,n-r)  →  Doğrulandı")
print(f"   Örnek: C(9,3) = {kombinasyon(9,3)} = C(9,6) = {kombinasyon(9,6)}")

# Özellik 3: Pascal Kuralı
for n in range(2, 10):
    for r in range(1, n+1):
        assert kombinasyon(n,r) == kombinasyon(n-1,r) + kombinasyon(n-1,r-1)
print(f"✓ Özellik 3 (Pascal): C(n,r)=C(n-1,r)+C(n-1,r-1)  →  Doğrulandı")
print(f"   Örnek: C(6,3) = C(5,3)+C(5,2) = {kombinasyon(5,3)}+{kombinasyon(5,2)} = {kombinasyon(6,3)}")
```

```python
# Pascal Üçgeni Görselleştirme
satirlar = 9
fig, ax = plt.subplots(figsize=(11, 6))
ax.set_xlim(-0.5, satirlar - 0.5)
ax.set_ylim(-0.5, satirlar - 0.5)
ax.axis('off')
ax.set_title("Pascal Üçgeni — C(n,r) değerleri", fontsize=13, fontweight='bold')

max_val = kombinasyon(satirlar - 1, (satirlar - 1) // 2)
cmap = plt.cm.Blues

for n in range(satirlar):
    for r in range(n + 1):
        val = kombinasyon(n, r)
        x = r - n / 2 + (satirlar - 1) / 2
        y = satirlar - 1 - n
        intensity = 0.15 + 0.75 * (val / max_val) ** 0.4
        circle = plt.Circle((x, y), 0.42, color=cmap(intensity), zorder=2)
        ax.add_patch(circle)
        fontsize = 10 if val < 1000 else 7
        ax.text(x, y, str(val), ha='center', va='center',
                fontsize=fontsize, fontweight='bold',
                color='white' if intensity > 0.5 else 'black', zorder=3)
        ax.text(x - 0.5, y + 0.28, f"C({n},{r})",
                ha='center', va='center', fontsize=5, color='gray', zorder=3)

plt.tight_layout()
plt.show()
print("Her satırda komşu iki değerin toplamı bir alt satırı verir (Pascal Kuralı).")
```

### Binom Teoremi
$$
(a+b)^n = \sum_{k=0}^{n} \binom{n}{k} a^k b^{n-k}
$$

```python

a, b = sp.symbols('a b')

print("Binom Teoremi — Açılımlar:")
print("-" * 45)
for n in range(1, 7):
    acilim = sp.expand((a + b)**n)
    acilim_str = str(acilim).replace('**', '^').replace('*', '·')
    print(f"(a+b)^{n} = {acilim_str}")

# Sayısal doğrulama
print("\nSayısal Doğrulama:")
for n in [3, 5, 8]:
    # (a+b)^n = sum C(n,k) a^k b^(n-k)
    a_val, b_val = 2, 3
    direkt = (a_val + b_val) ** n
    toplam = sum(kombinasyon(n, k) * a_val**k * b_val**(n-k) for k in range(n+1))
    print(f"  ({a_val}+{b_val})^{n} = {direkt} = {toplam}  {'✓' if direkt==toplam else '✗'}")
```

## 4. Çok Terimli Katsayılar

$n$ nesneyi $r$ gruba ayırma (grup boyutları: $n_1, n_2, \ldots, n_r$, toplam $n$):

$$\binom{n}{n_1, n_2, \ldots, n_r} = \frac{n!}{n_1!\, n_2!\, \cdots\, n_r!}$$

Çok Terimli Teorem:
$$(a_1 + a_2 + \cdots + a_r)^n = \sum_{n_1+\cdots+n_r=n} \binom{n}{n_1, n_2, \ldots, n_r} a_1^{n_1} a_2^{n_2} \cdots a_r^{n_r}$$

```python
def cok_terimli(n, gruplar):
    """Çok terimli katsayı: n! / (n1! * n2! * ... * nr!)"""
    assert sum(gruplar) == n, "Grup boyutları toplamı n'e eşit olmalı!"
    payda = math.prod(faktoriyel(g) for g in gruplar)
    return faktoriyel(n) // payda

print("Çok Terimli Katsayı Örnekleri:")
print("-" * 55)

# Polis departmanı örneği
n = 10
gruplar = [5, 2, 3]
sonuc = cok_terimli(n, gruplar)
print(f"Örnek: 10 memur → [devriye:{gruplar[0]}, evrak:{gruplar[1]}, dükkan:{gruplar[2]}]")
print(f"  C(10; 5,2,3) = 10! / (5!·2!·3!) = {faktoriyel(10)} / ({faktoriyel(5)}·{faktoriyel(2)}·{faktoriyel(3)})")
print(f"             = {sonuc:,}")

print()
# Diğer örnekler
ornekler = [
    (6,  [2, 2, 2], "6 oyuncuyu 3 eşit takıma"),
    (12, [4, 4, 4], "12 öğrenciyi 3 eşit gruba"),
    (9,  [3, 3, 3], "9 projeyi 3 gruba"),
]
for n, grp, aciklama in ornekler:
    print(f"{aciklama}: C({n}; {grp}) = {cok_terimli(n, grp):,}")

# r=2 özel durumu: binom katsayısı
print("\nÖzel durum r=2 → Binom katsayısı:")
for n in [5, 8, 10]:
    r1 = n // 2
    r2 = n - r1
    ct = cok_terimli(n, [r1, r2])
    binom = kombinasyon(n, r1)
    print(f"  C({n}; {r1},{r2}) = {ct} = C({n},{r1}) = {binom}  {'✓' if ct==binom else '✗'}")
```

```python
# Görselleştirme: Polis departmanı örneği
n_memur = 10
gorev_isimleri = ['Devriye (5)', 'Evrak (2)', 'Donut Dükkanı (3)']
gorev_sayilari = [5, 2, 3]
gorev_renkleri = ['#4C72B0', '#DD8452', '#55A868']

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Sol: Pasta grafik
axes[0].pie(gorev_sayilari, labels=gorev_isimleri, colors=gorev_renkleri,
            autopct=lambda p: f'{p:.0f}%\n({round(p*n_memur/100)} kişi)',
            startangle=90, textprops={'fontsize': 11},
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[0].set_title(f'10 Memur — Görev Dağılımı\nC(10;5,2,3) = {cok_terimli(10,[5,2,3]):,} yol',
                  fontsize=12, fontweight='bold')

# Sağ: Çok terimli vs binom karşılaştırma
ax2 = axes[1]
ns2 = range(3, 13)
# C(n, n//3, n//3, n - 2*(n//3)) yaklaşık eşit 3 grup
vals_3 = []
vals_2 = []
for n2 in ns2:
    g1 = n2 // 3
    g2 = n2 // 3
    g3 = n2 - g1 - g2
    if g1 > 0 and g2 > 0 and g3 > 0:
        vals_3.append(cok_terimli(n2, [g1, g2, g3]))
    else:
        vals_3.append(0)
    vals_2.append(kombinasyon(n2, n2 // 2))

x2 = np.array(list(ns2))
ax2.semilogy(x2, vals_2, 'o-', label='C(n, n/2) — 2 grup', color='#4C72B0', linewidth=2)
x2_filt = x2[[i for i, v in enumerate(vals_3) if v > 0]]
ax2.semilogy(x2_filt, [v for v in vals_3 if v > 0],
             's--', label='C(n; n/3,n/3,n/3) — 3 grup',
             color='#DD8452', linewidth=2,
             markersize=7)
ax2.set_xlabel('n (toplam nesne)', fontsize=11)
ax2.set_ylabel('Bölünme Sayısı (log ölçeği)', fontsize=11)
ax2.set_title('Binom vs Çok Terimli Büyüme', fontsize=12, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
